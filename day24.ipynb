```perl
sub do_wire {
    my ($op, $x, $y) = @_;
    return $op eq 'OR' ? $x | $y :
        $op eq 'XOR' ? $x ^ $y :
        $x & $y;
}

sub process_day24_part1 {
    my ($input) = @_;
    my ($init, $wires) = split "\n\n", $input;
    my %data = map { length == 1 ? int : $_ } $init =~ /(\w\d+): (\d+)/g;
    my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my @solvable =
        grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
        @wire_instructions;
    while (@solvable) {
        my @next_instructions =
            grep { not (exists $data{$_->[0]} and exists $data{$_->[2]}) }
            @wire_instructions;
        @data{map $_->[3], @solvable} =
            map do_wire($_->[1], @data{@{$_}[0, 2]}),
            @solvable;
        @wire_instructions = @next_instructions;
        @solvable =
            grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
            @wire_instructions;
        timecheck 5;
    }

    oct "0b" .
        join '',
        reverse
        map $data{$_},
        sort
        grep /\Az/,
        keys %data
}

timer sub {
    say Dumper process_day24_part1($practice_input);
    # confirm process_day24_part1($practice_input), 7;
}
```

```perl
sub calc_wiremap {
    my ($init_data, @wire_instructions) = @_;
    # my ($init, $wires) = split "\n\n", $input;
    # my %data = map { length == 1 ? int : $_ } $init =~ /(\w\d+): (\d+)/g;
    # my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my %data = %$init_data;
    my @solvable =
        grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
        @wire_instructions;
    while (@solvable) {
        my @next_instructions =
            grep { not (exists $data{$_->[0]} and exists $data{$_->[2]}) }
            @wire_instructions;
        @data{map $_->[3], @solvable} =
            map {
                my ($op, $x, $y) = ($_->[1], @data{@{$_}[0, 2]});
                $op eq 'OR' ? $x | $y :
                    $op eq 'XOR' ? $x ^ $y :
                    $x & $y;
                # do_wire($_->[1], @data{@{$_}[0, 2]})
            }
            @solvable;
        @wire_instructions = @next_instructions;
        @solvable =
            grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
            @wire_instructions;
    }

    
        join '',
        reverse
        map $data{$_},
        sort
        grep /\Az/,
        keys %data
}


sub do_wire2 {
    my ($op, $x, $y) = @_;
    return $op eq 'OR' ? "($x | $y)" :
        $op eq 'XOR' ? "($x ^ $y)" :
        "($x & $y)";
}

sub string_map_wires {
    my ($input) = @_;
    my ($init, $wires) = split "\n\n", $input;
    my %data = map { $_ => $_ } $init =~ /(\w\d+): \d+/g;
    my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my @solvable =
        grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
        @wire_instructions;
    while (@solvable) {
        my @next_instructions =
            grep { not (exists $data{$_->[0]} and exists $data{$_->[2]}) }
            @wire_instructions;
        @data{map $_->[3], @solvable} =
            map do_wire2($_->[1], @data{@{$_}[0, 2]}),
            @solvable;
        @wire_instructions = @next_instructions;
        @solvable =
            grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
            @wire_instructions;
        timecheck 5;
    }

        map $data{$_},
        sort
        grep /\Az/,
        keys %data;
    # \%data
}
```

```perl
sub random_bitstring { join '', map int(2 * rand), 0 .. $_[0] - 1 }
sub test_incorrect_bits {
    my ($input_length, $expc_fun, $test_x, $test_y, @wire_instructions) = @_;
    my %data;
    my $expc_z = $expc_fun->($test_x, $test_y);
    # map { $_ => } map sprintf("x%02d", $_), 0 .. $input_length - 1;
    %data = ((map { sprintf("x%02d", $_) => int substr $test_x, -1-$_, 1 } 0 .. $input_length - 1),
        (map { sprintf("y%02d", $_) => int substr $test_y, -1-$_, 1 } 0 .. $input_length - 1));
    my $test_r = calc_wiremap(\%data, @wire_instructions);
    # die "wat: $test_x & $test_y -> '$expc_z' vs '$test_r'" if (($expc_z ^ $test_r) | ('0' x $input_length)) =~ /[^01]/;
    return [$test_x, $test_y, $expc_z, $test_r, $expc_z ^ $test_r | '0' x $input_length, earliest_incorrect_bit(($expc_z ^ $test_r) | ('0' x $input_length)) ] if $expc_z ne $test_r;
    return undef;
}

sub sample_incorrect_bits {
    my ($input_length, $expc_fun, @wire_instructions) = @_;
    my $test_x = random_bitstring($input_length);
    my $test_y = random_bitstring($input_length);
    return test_incorrect_bits($input_length, $expc_fun, $test_x, $test_y, @wire_instructions);
}

sub find_incorrect_samples {
    grep defined $_->[5],
    grep defined,
    map sample_incorrect_bits(@_),
    0 .. 250
}

sub retest_incorrect_samples {
    my ($input_length, $expc_fun, $samples, @wire_instructions) = @_;
    grep defined ($_->[5]),
    grep defined,
    map test_incorrect_bits($input_length, $expc_fun, $_->[0], $_->[1], @wire_instructions),
    @$samples
}


sub quick_incorrect_samples {
    my ($input_length, $expc_fun, $target_bit, $samples, @wire_instructions) = @_;

    foreach (@$samples) {
        my $res = test_incorrect_bits($input_length, $expc_fun, $_->[0], $_->[1], @wire_instructions);
        next unless defined $res;
        next unless defined $res->[5];
        # appendfile(".exp/day24_test_wires", "got possible incorrect sample res -> " . (Dumper $res) . "\n");
        return $res->[5] if $res->[5] <= $target_bit;
    }
    return;
}

sub earliest_incorrect_bit {
    my ($error_bits) = @_;
    my @c = reverse split '', $error_bits;
    foreach (0 .. $#c) {
        # confess "invalid error bits: '$error_bits'" if $c[$_] !~ /\d/;
        return $_ if $c[$_] == 1;
    }
    return;
}
```

```perl

sub do_wire3 {
    my ($op, @inputs) = @_;
    return join ',', @inputs;
}

sub find_swappable_bits {
    my ($input_length, @wire_instructions) = @_;
    my %data = ((map { sprintf("x%02d", $_) => sprintf("x%02d", $_) } 0 .. $input_length - 1),
        (map { sprintf("y%02d", $_) => sprintf("y%02d", $_) } 0 .. $input_length - 1));
    # my ($input) = @_;
    # my ($init, $wires) = split "\n\n", $input;
    # my %data = map { $_ => $_ } $init =~ /(\w\d+): \d+/g;
    # my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my @solvable =
        grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
        @wire_instructions;
    while (@solvable) {
        my @next_instructions =
            grep { not (exists $data{$_->[0]} and exists $data{$_->[2]}) }
            @wire_instructions;
        @data{map $_->[3], @solvable} =
            map do_wire3($_->[1], @{$_}[0, 2], @data{@{$_}[0, 2]}),
            @solvable;
        @wire_instructions = @next_instructions;
        @solvable =
            grep { exists $data{$_->[0]} and exists $data{$_->[2]} }
            @wire_instructions;
        # timecheck 5;
    }

    return {
        map { $_ => { map { $_ => 1 } $_, grep $_ !~ /\A[xy]\d+\Z/, sort split ',', $data{$_} } }
        sort
        grep /\Az/,
        keys %data };
    
    # \%data
}

sub inspect_find_swappable_bits {
    my ($input) = @_;
    my ($init, $wires) = split "\n\n", $input;
    my %data = map { length == 1 ? int : $_ } $init =~ /(\w\d+): (\d+)/g;
    my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my $input_length = scalar grep /\Ax/, keys %data;
    return find_swappable_bits($input_length, @wire_instructions);
}

sub do_swap_instructions {
    my ($x, $y, @wire_instructions) = @_;
    my ($xindex) = grep $wire_instructions[$_][3] eq $x, 0 .. $#wire_instructions;
    my ($yindex) = grep $wire_instructions[$_][3] eq $y, 0 .. $#wire_instructions;
    do_swap_instructions_numeric($xindex, $yindex, @wire_instructions);
}
sub do_swap_instructions_numeric {
    my ($i, $j, @wire_instructions) = @_;

    my @inst = @wire_instructions;
    my @swaps = map [ @$_ ], @inst[$i, $j];
    my @labels = reverse map $_->[3], @swaps;
    $swaps[$_][3] = $labels[$_] foreach 0,1;
    @inst[$i, $j] = @swaps;

    return @inst;
}

timer sub {
    my $input = get_challenge('2024/day/24/input');
    say Dumper inspect_find_swappable_bits($input);
    # confirm $res, 1151;
};
```

```perl
sub search_for_wire_fix {
    my ($input_length, $expc_fun, $swappable_bits, @wire_instructions) = @_;
    my @samples = sort { $a->[5] <=> $b->[5] } find_incorrect_samples($input_length, $expc_fun, @wire_instructions);
    return unless @samples;
    my $target_bit = $samples[0][5];
    unless (defined $target_bit) {
        say "found no working samples!";
        say Dumper @samples;
        return;
    }
    say "target bit: $target_bit";
    appendfile(".exp/day24_test_wires", "target bit: $target_bit" . "\n");

    my @target_samples = grep { $_->[5] == $target_bit } @samples;
    # @target_samples = @target_samples[0 .. 19] if @target_samples > 20;
    appendfile(".exp/day24_test_wires", "target samples: " . Dumper(@target_samples[0 .. 2]) . "\n");

    my %relevant_instructions = %{$swappable_bits->{sprintf "z%02d", $target_bit}};
    my %previous_instructions =
        map %$_,
        map $swappable_bits->{sprintf "z%02d", $_}, 0 .. $target_bit-1;
    my @search_space_i = grep { exists $relevant_instructions{$wire_instructions[$_][3]} } 0 .. $#wire_instructions;
    my @search_space_j = grep { not exists $previous_instructions{$wire_instructions[$_][3]} } 0 .. $#wire_instructions;
    appendfile(".exp/day24_test_wires", "search space: " . scalar(@search_space_i) . ' x ' . scalar(@search_space_j) . "\n");
    my $counter = 0;
    foreach my $i (@search_space_i) {
        # say "testing i: $i";
        foreach my $j (@search_space_j) {
            next if $i == $j;
            my @inst = do_swap_instructions_numeric($i, $j, @wire_instructions);
            my @labels = map $_->[3], @inst[$i, $j];
            # my @swaps = map [ @$_ ], @inst[$i, $j];
            # my @labels = reverse map $_->[3], @swaps;
            # $swaps[$_][3] = $labels[$_] foreach 0,1;
            # @inst[$i, $j] = @swaps;

            my $res = quick_incorrect_samples($input_length, $expc_fun, $target_bit, \@target_samples, @inst);
            # say "testing swap: $i <=> $j -> $res";
            appendfile(".exp/day24_test_wires", "testing swap: $i <=> $j ($labels[0] <=> $labels[1]) -> " . ($res // '') . "\n") if $counter++ % 200 == 0;
            if ((not defined $res) or $res > $target_bit+1) {
                appendfile(".exp/day24_test_wires", "found possible: $i <=> $j ($labels[0] <=> $labels[1]) -> " . ($res // '') . "\n");
                my @more_samples = grep { $_->[5] <= $target_bit } find_incorrect_samples($input_length, $expc_fun, @inst);
                $res = quick_incorrect_samples($input_length, $expc_fun, $target_bit, \@more_samples, @inst);
                if ((not defined $res) or $res > $target_bit+1) {
                    appendfile(".exp/day24_test_wires", "found confident solution: $i <=> $j ($labels[0] <=> $labels[1]) -> " . ($res // '') . "\n");
                    say "found solution to fix [$target_bit]: $i <=> $j ($labels[0] <=> $labels[1])";
                    return $i, $j;
                } else {
                    appendfile(".exp/day24_test_wires", "false find, adding more samples to dataset\n");
                    push @target_samples, @more_samples;
                }
                    
            }
            timecheck 60 * 4;
        }
    }
    return;
}
```

```perl


sub process_day24_part2 {
    my ($input, $expc_fun, @previous_swaps) = @_;
    my ($init, $wires) = split "\n\n", $input;
    my %data = map { length == 1 ? int : $_ } $init =~ /(\w\d+): (\d+)/g;
    my @wire_instructions = map { [ /(\S+) (OR|XOR|AND) (\S+) -> (\S+)/ ] } split "\n", $wires;
    my $input_length = scalar grep /\Ax/, keys %data;

    # apply any previous swaps
    foreach my $prev (@previous_swaps) {
        say "pre-swapping: ", join ',', @$prev;
        @wire_instructions = do_swap_instructions(@$prev, @wire_instructions);
    }
    # say Dumper map { join ' -> ', @$_ } @wire_instructions;
    # return;

    # calculate helpers
    my $swappable_bits = find_swappable_bits($input_length, @wire_instructions);
    # search for fix
    my ($i, $j) = search_for_wire_fix($input_length, $expc_fun, $swappable_bits, @wire_instructions);

    while (defined $i) {
        # perform swap
        @wire_instructions = do_swap_instructions_numeric($i, $j, @wire_instructions);
        # calculate helpers
        $swappable_bits = find_swappable_bits($input_length, @wire_instructions);
        # search for fix
        ($i, $j) = search_for_wire_fix($input_length, $expc_fun, $swappable_bits, @wire_instructions);
    }

}


timer sub {
    # say Dumper process_day24_part2($practice_input);
    say Dumper process_day24_part2($and_test, sub { $_[0] & $_[1] });
    say Dumper process_day24_part2($and_test, sub { $_[0] & $_[1] }, ['z05', 'z00']);
    # say Dumper process_day24_part2($practice_input);
    # confirm process_day24_part2($practice_input), 7;
}
```
