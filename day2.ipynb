```perl
$graphql::graphql_methods{is_valid_level} = sub {
        my ($self, $params) = @_;
        my $value = abs($self->{value} // 0);
        return (1 <= $value && $value <= 3 ? 1 : 0);
    };
$graphql::graphql_methods{is_valid_sign} = sub {
        my ($self, $params) = @_;
        my $self_sign = ($self->{value} // 0) >= 0 ? 1 : -1;
        my $root_sign = $self->{arr}[$self->{coord}[0]][1] >= 0 ? 1 : -1;
        return $self_sign == $root_sign;
    };
$graphql::graphql_methods{is_first} = sub {
        my ($self, $params) = @_;
        return $self->{coord}[1] == 0;
    };
$graphql::graphql_methods{is_valid} = sub {
        my ($self, $params) = @_;
        return $self->{is_first}($self) || ($self->{is_valid_level}($self) && $self->{is_valid_sign}($self));
    };

# say Dumper \%graphql::graphql_methods;

# say Dumper slice_nd([[], []],
#     map_nd selector 'is_valid',
#     map_nd_indexed graphql_query("{ is_valid }"),
#     map_nd selector 'delta.value',
#     map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
#     parse_2d_string_array($practice_input));
# sum map { all { $_ } @$_ } @{
#     map_nd selector 'is_valid',
#     map_nd_indexed graphql_query("{ is_valid }"),
#     map_nd selector 'delta.value',
#     map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
#     parse_2d_string_array($practice_input)};
```

```perl
$graphql::graphql_methods{where} = sub {
        my ($self, $params) = @_;
        return $self->{value} == $params->{value} ? $self : undef;
    };

sub remove_indices {
    my ($arr, @indices) = @_;
    foreach my $index (@indices) {
        splice @{$arr->[$index->[0]]}, $index->[1], 1;
    }
    return $arr;
}

my @removable =
    grep defined,
    map { first { defined } @$_ } @{
    map_nd selector 'where.coord',
    map_nd_indexed graphql_query("{ where(value:0) { coord } }"),
    map_nd sub { $_->{is_first} || ($_->{is_valid_level} && $_->{is_valid_sign}) },
    map_nd_indexed graphql_query("{ is_first is_valid_level is_valid_sign }"),
    map_nd selector 'delta.value',
    map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
    parse_2d_string_array($practice_input)};

# say Dumper remove_indices(parse_2d_string_array($practice_input), @removable);


sum map { all { $_ } @$_ } @{
    map_nd sub { $_->{is_first} || ($_->{is_valid_level} && $_->{is_valid_sign}) },
    map_nd_indexed graphql_query("{ is_first is_valid_level is_valid_sign }"),
    map_nd selector 'delta.value',
    map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
    remove_indices parse_2d_string_array($practice_input), @removable };
```

```perl
$graphql::graphql_methods{is_valid_first} = sub {
        my ($self, $params) = @_;
        my @row = @{$self->{arr}[$self->{coord}[0]]};
        return $self->{coord}[1] > 0 ? $self->{value} : ($self->{value} && !(none { $_ } @row[2 .. $#row]));
    };
$graphql::graphql_methods{is_high_jump_first} = sub {
        my ($self, $params) = @_;
        my @row = @{$self->{arr}[$self->{coord}[0]]};
        say "found: ", $self->{coord}[0] if ($self->{coord}[1] == 0 && abs($row[1]) > 3 && abs($row[2]) <= 3);
        return ($self->{coord}[1] == 0 && abs($row[1]) > 3 && abs($row[2]) <= 3);
    };
say Dumper
    map { join ' ', map int, @$_ } 
    @{
        map_nd selector 'is_valid_first',
        map_nd_indexed graphql_query("{ is_valid_first }"),
        map_nd sub { $_->{is_first} ? !$_->{is_high_jump_first} : $_->{is_valid_level} && $_->{is_valid_sign} },
        map_nd_indexed graphql_query("{ is_high_jump_first is_first is_valid_level is_valid_sign }"),
        map_nd selector 'delta.value',
        map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
        slice_nd [[undef, 643]],
            parse_2d_string_array($input)};
my @removable_initial =
    map { first { defined } @$_ }
    @{
            map_nd selector 'where.coord',
            map_nd_indexed graphql_query("{ where(value:0) { coord } }"),
            map_nd selector 'is_valid_first',
            map_nd_indexed graphql_query("{ is_valid_first }"),
            map_nd sub { $_->{is_first} ? !$_->{is_high_jump_first} : $_->{is_valid_level} && $_->{is_valid_sign} },
            map_nd_indexed graphql_query("{ is_high_jump_first is_first is_valid_level is_valid_sign }"),
            map_nd selector 'delta.value',
            map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
            slice_nd [[undef, 643]],
                parse_2d_string_array($input)};
```

```perl
# say 'fixable?: ', Dumper
my @fixable = map { first { defined } @$_ }
    @{
        map_nd selector 'where.pos.a',
        map_nd_indexed graphql_query("{ where(value:0) { pos { a } } }"),
        map_nd sub { $_->{is_first} || ($_->{is_valid_level} && $_->{is_valid_sign}) },
        map_nd_indexed graphql_query("{ is_first is_valid_level is_valid_sign }"),
        map_nd selector 'delta.value',
        map_nd_indexed graphql_query("{ delta(db:-1) { value } }"),
        $data_initial_removable };
say Dumper @fixable
```

```perl
my @fixed = grep { not defined $fixable[$_] } 0 .. 643;
my %fixed_map = map { $_ => 1 } @fixed;
say 'fixed: ', Dumper \@fixed;
```

```perl
sub is_correct {
    my ($input) = @_;
    all { $_ } @{(map_nd sub { $_->{is_first} || ($_->{is_valid_level} && $_->{is_valid_sign}) },
        map_nd_indexed graphql_query("{ is_first is_valid_level is_valid_sign }"),
        map_nd selector 'delta.value',
        map_nd_indexed graphql_query("{ delta(db:-1) { value } }"), [ $input ])->[0]};
}

sub reslice_and_correct {
    my ($input) = @_;
    return 1 if is_correct($input);
    foreach my $i (0 .. $#{$input}) {
        my @copy = @$input;
        splice @copy, $i, 1;
        if (is_correct(\@copy)) {
            say "fixed as: ", join ' ', @copy;
            return 1;
        }
    }
    return 0;
}

say Dumper is_correct([ 1 .. 5, 3 ]);
say Dumper reslice_and_correct([ 0, 1 .. 5, 6 .. 10 ]);
```

```perl
my $data = parse_2d_string_array($input);
$data = slice_nd [[0, 643]], $data;
my $counter = 0;
for my $i (0 .. 643) {
    # next if exists $fixed_map{$i};
    if (reslice_and_correct($data->[$i])) {
        say "fixable: $i";
        $counter++;
    }
}
$counter
```
