```perl
sub shuffle { map $_->[1], sort { $a->[0] <=> $b->[0] } map [rand, $_], @_ }
```

```perl
sub process_day13_part1 {
    my ($input) = @_;
    my @puzzles = split "\n\n", $input;
    
    my @solutions = map {
            my $problem = $_;
            die "wat: $_" unless /Button A: X\+(\d+), Y\+(\d+)/;
            my $pa = [int $1, int $2];
            die "wat: $_" unless /Button B: X\+(\d+), Y\+(\d+)/;
            my $pb = [int $1, int $2];
            die "wat: $_" unless /Prize: X=(\d+), Y=(\d+)/;
            my $target = [int $1, int $2];
    
            my @possible_solutions =
                sort { $a->[2] <=> $b->[2] }
                map { [ @$_, $_->[0] * 3 + $_->[1] ] }
                grep { $_->[0] >= 0 && $_->[1] >= 0 && $_->[0] <= 100 && $_->[1] <= 100 }
                map { [ $_, ($target->[0] - $pa->[0] * $_) / $pb->[0] ] }
                grep { (($target->[0] - $pa->[0] * $_) / $pb->[0]) == (($target->[1] - $pa->[1] * $_) / $pb->[1]) }
                grep { ($target->[1] - $pa->[1] * $_) % $pb->[1] == 0 }
                grep { ($target->[0] - $pa->[0] * $_) % $pb->[0] == 0 }
                # map { ($target->[0] - $pa->[0] * $_) % $b->[0] }
                0 .. int($target->[0] / $pa->[0])
            ;
            # say "cool: $problem", Dumper @possible_solutions if @possible_solutions > 1;
            die if
                grep { !(($_->[0] * $pa->[0] + $_->[1] * $pb->[0]) == $target->[0] && ($_->[0] * $pa->[1] + $_->[1] * $pb->[1]) == $target->[1] ) }
                @possible_solutions;
    
            @possible_solutions
                ? [ $possible_solutions[0],
                    "$problem\n > solution: $possible_solutions[0][0] * A, $possible_solutions[0][1] * B (costing $possible_solutions[0][2])"
                    . ( @possible_solutions > 1 ? "\n    other possible solutions:\n" . join("\n", map "$_->[0] * A, $_->[1] * B (cost $_->[2])", @possible_solutions[1 .. $#possible_solutions]) : '') ]
                : [ undef, "$problem\n > no solution" ]
        }
        flatten_nd
        # map_nd_indexed sub { my ($arr, $p) = @_; "id: $p->[0]\n" . $arr->[$p->[0]] },
        [ @puzzles ];

    my @debug_data = map $_->[1], @solutions;
    say join "\n", @debug_data;

    sum
        map $_->[0][2],
        grep defined $_->[0],
        @solutions
}
timer sub {
    say Dumper process_day13_part1($practice_input);
    confirm process_day13_part1($practice_input), 480;
}
```

```perl
my $res;
timer sub {
    my $input = get_challenge('2024/day/13/input');
    $res = process_day13_part1($input);
    confirm $res, 39996;
};
say Dumper $res;
```

```perl

sub round { sprintf "%.0f", $_[0] }

sub solve_with_rref {
    my ($problem) = @_;
    die "wat: $_" unless $problem =~ /Button A: X\+(\d+), Y\+(\d+)/;
    my $pa = [int $1, int $2];
    die "wat: $_" unless $problem =~ /Button B: X\+(\d+), Y\+(\d+)/;
    my $pb = [int $1, int $2];
    die "wat: $_" unless $problem =~ /Prize: X=(\d+), Y=(\d+)/;
    my $target = [int $1, int $2];
    
    my $rref = [
     [$pa->[0], $pb->[0], $target->[0]],
     [$pa->[1], $pb->[1], $target->[1]]
    ];
    
    my @shape = shape $rref;
    my $row = 0;
    my $div = $rref->[$row][$row];
    $rref->[$row] = [ map $rref->[$row][$_] / $div, 0 .. $shape[1] - 1 ];
    
    $row = 1;
    my $factor = $rref->[$row][0];
    $rref->[$row] = [ map $rref->[$row][$_] - $factor * $rref->[0][$_], 0 .. $shape[1] - 1 ];
    
    $div = $rref->[$row][$row];
    $rref->[$row] = [ map $rref->[$row][$_] / $div, 0 .. $shape[1] - 1 ];
    
    $rref->[0][2] -= $rref->[0][1] * $rref->[1][2];
    $rref->[0][1] -= $rref->[0][1];
    $rref->[0][2] += 0;
    $rref->[1][2] = $rref->[1][2] + 0;
    
    # warn "rref result: ", Dumper $rref;
    say Dumper $rref;
    say "error margin:", Dumper [ map abs($_->[2] - round $_->[2]), @$rref ];
    return undef unless all sub { abs($_->[2] - round $_->[2]) < 0.001 }, @$rref;
    

    return [ map $_->[2], @$rref ];
}

# say Dumper solve_with_rref(qq/
# Button A: X+94, Y+34
# Button B: X+22, Y+67
# Prize: X=10000000008400, Y=10000000005400
# /);

say Dumper solve_with_rref(qq/
Button A: X+56, Y+17
Button B: X+33, Y+61
Prize: X=5469, Y=5178
/);
```

```perl
sub increase_problem {
    my ($problem) = @_;
    $problem =~ /Prize: X=(\d+), Y=(\d+)/;
    my $x = $1 + 10000000000000;
    my $y = $2 + 10000000000000;
    return $problem =~ s/Prize: X=(\d+), Y=(\d+)/Prize: X=$x, Y=$y/r;
}
sub process_day13_part2 {
    my ($input, $increase) = @_;
    my @puzzles = split "\n\n", $input;
    
    my @solutions = map {
            my $problem = $_;
            $problem = increase_problem($problem) if $increase;
            my @possible_solutions =
                sort { $a->[2] <=> $b->[2] }
                map { [ @$_, $_->[0] * 3 + $_->[1] ] }
                grep defined,
                solve_with_rref($problem);
            ;
            # die if
            #     grep { !(($_->[0] * $pa->[0] + $_->[1] * $pb->[0]) == $target->[0] && ($_->[0] * $pa->[1] + $_->[1] * $pb->[1]) == $target->[1] ) }
            #     @possible_solutions;
    
            @possible_solutions
                ? [ $possible_solutions[0],
                    "$problem\n > solution: A=$possible_solutions[0][0], B=$possible_solutions[0][1] (costing $possible_solutions[0][2])" ]
                : [ undef, "$problem\n > no solution" ]
        }
        flatten_nd
        # map_nd_indexed sub { my ($arr, $p) = @_; "id: $p->[0]\n" . $arr->[$p->[0]] },
        [ @puzzles ];

    my @debug_data = map $_->[1], @solutions;
    say join "\n", @debug_data;

    sum
        map $_->[0][2],
        grep defined $_->[0],
        @solutions
}
timer sub {
    # say Dumper process_day13_part2($practice_input);
    confirm process_day13_part2($practice_input), 480;
    my $res;
    my $input = get_challenge('2024/day/13/input');
    $res = process_day13_part2($input);
    confirm $res, 39996;
};
```
